```python
import random
import datetime
l=[] #
info={} #Stores the account information of the customer
classes={} # Stores the instances of the bank acoounts
class Bank:
    global l
    
    def __init__(self,accountNo=0,savings=0):
        self.accountNo = accountNo
        self.savings = savings
        self.n = 1
        self.d = {}
    def AccountNoGen(self,l):
        i = random.randint(10**9,10**10-1)
        while i in l:
            i = random.randint(10**9,10**10-1)
        l.append(i)
        self.accountNo = i
        self.n+=1
        self.d[datetime.datetime.now()] = "account created with no: "+str(self.accountNo)
    def Deposit(self,money):
        self.savings = self.savings + money
        self.d[datetime.datetime.now()] = "Amount of money deposited: "+str(money)
        self.n+=1
    def withdraw(self,money):
        self.savings = self.savings - money
        self.d[datetime.datetime.now()] = "Amount of money withdrawn: "+str(money)
        self.n+=1
    def getBalance(self):
        print(self.savings)
    def getLog(self):
        for z in self.d:
            print(z,":",self.d[z])
    def totalTransactions(self):
        print("total number of operations done on this account: ",self.n)
    def terminateAccount(self):
        l.remove(self.accountNo)
        self.d={}
        self.n = 1    
```

```python
while True:
    global info,l,classes
    i = int(input("Enter Your Choice \n 1 : create account \n 2 : deposit money \n 3 : withdraw money \n 4 : get the log details \n 5 : know total number of transactions \n 6 : get your balance \n -1 : end"))
    if i==1:
        name_ref=name = input("Enter your full name:")
        name = Bank()
        classes[name_ref]=name
        name.AccountNoGen(l)
        info[name.accountNo]=name_ref
        print(name.accountNo)
        print(info)
    elif i==2:
        acc_no=int(input("Enter Account Number:"))
        try:
            name=classes[info[acc_no]]
            money=int(input("Enter Amount of money to be deposited:"))
            name.Deposit(money)
        except:
            print("Enter a valid account number")        
    elif i==3:
        acc_no=int(input("Enter Account Number:"))
        try:
            name=classes[info[acc_no]]
            money=int(input("Enter amount of money to be withdrawn:"))
            name.withdraw(money)
        except:
            print("Enter a valid account number")   
    elif i==4:
        acc_no=int(input("Enter Account Number:"))
        try:
            name=classes[info[acc_no]]
            name.getLog()
        except:
            print("Enter a valid account number") 
    elif i==5:
        acc_no=int(input("enter account number:"))
        try:
            name=classes[info[acc_no]]
            name.totalTransactions()
        except:
            print("Enter a valid account number") 
    elif i==6:
        acc_no=int(input("enter account number:"))
        try:
            name=classes[info[acc_no]]
            name.getBalance()
        except:
            print("Enter a valid account number") 
        
    else:
        print("Thank You For Banking")
        break
```
